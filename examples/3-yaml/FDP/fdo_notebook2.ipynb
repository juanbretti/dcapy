```python
oil_mr = MeanReversion(
    initial_condition = 66,
    ti = date(2021,1,1),
    generator = {'dist':'norm','kw':{'loc':0,'scale':5.13}},
    m=46.77,
    eta=0.112652,
    freq = 'A'
)
```

```python
price_mr = oil_mr.generate(12,50, freq_output='A', seed=21)

price_mr.plot(legend=False)
```
